```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the data
data = pd.read_csv(r"C:\Users\sruja\Downloads\student_kmeans_dataset.csv")

# Select features
X = data[['Study_Hours', 'Exam_Score']]

# Apply K-means with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

# Plot clusters
plt.scatter(data['Study_Hours'], data['Exam_Score'], c=data['Cluster'], cmap='viridis', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', s=200)
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('K-Means Clustering (K=3)')
plt.show()
```
